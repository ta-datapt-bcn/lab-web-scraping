#### Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

#### Display all the image links from Walt Disney wikipedia page

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
#your code
links_disney = requests.get(url).content
html_disney = BeautifulSoup(links_disney, "lxml")
images_disney = html_disney.find_all('a',{'class':'image'})
images_disney
````

````python
images_disney = [row.get('href') for row in images_disney]
images_disney
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
````

````python
#your code
python = requests.get(url).content
pythonsoup = BeautifulSoup(python, "lxml")
python_links = pythonsoup.find_all('a')
python_links
````

````python
python_links2 = [row.get("href") for row in python_links]
python_links2
#necesitamos aquellas rutas que empiecen con http para filtrar solo links
````

````python
links_cleaned = [row for row in python_links2 if str(row).startswith("http")]
links_cleaned
````

#### Number of Titles that have changed in the United States Code since its last release point 

````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'
````

````python
#your code
usa_codes = requests.get(url).content
usa_codes_soup = BeautifulSoup(usa_codes, "lxml")
usa_titles = usa_codes_soup.find_all('div',{'class':'usctitlechanged'})
usa_titles
````

````python
usa_titles2 = [row.get_text() for row in usa_titles]
usa_titles2
````

````python
titles_clean = [row.replace('\n','') for row in usa_titles2]
titles_clean
````

````python
#recordatorio: usar .strip() para eliminar el espacio anterior al texto. arreglado en apartados anteriores
titles_clean2 = [row.strip() for row in titles_clean]
titles_clean2
````

#### A Python list with the top ten FBI's Most Wanted names 

````python
# This is the url you will scrape in this exercise
url = 'https://www.fbi.gov/wanted/topten'
````

````python
#your code 
fbi_mw = requests.get(url).content
fbi_mw_soup = BeautifulSoup(fbi_mw, "lxml")
fbi_names = fbi_mw_soup.find_all('h3',{'class':'title'})
fbi_names = [row.get_text() for row in fbi_names]
fbi_names
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
earthquakes = pd.read_html(url)
earthquakes_df = earthquakes[3]
earthquakes_df
````

````python
date = earthquakes_df['Date & Time UTC']
latitude = earthquakes_df['Latitude degrees']
longitude = earthquakes_df['Longitude degrees']
region_name = earthquakes_df['Region name [+]']
````

````python
latest20 = pd.concat([date, latitude, longitude, region_name],axis=1)
latest20
````

````python
top20 = latest20.iloc[:20]
top20
````

````python
columns = ['Date','Latitude position','Latitude direction','Longitude position','Longitude direction','Region']
top20.columns = columns
top20
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
#your code
from urllib.request import urlopen

html = urlopen(url)
wiki_soup = BeautifulSoup(html, 'html.parser')
langlist = wiki_soup.find_all('a',{'class':'link-box'})

for lang in langlist:
    print(lang.get_text())
````
