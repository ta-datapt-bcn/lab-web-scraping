#### Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
print(url)
````

````python
resp2=requests.get(url)
````

````python
resp2.status_code
````

````python
html2=resp2.text
soup2 = BeautifulSoup(html2,"html.parser")
repositories_list = soup2.findAll('h1', attrs={'class':"h3 lh-condensed"})
repositories_list
````

````python
repositories=[i.find(name='a').get('href').strip() for i in repositories_list]
repositories
````

#### Number of Titles that have changed in the United States Code since its last release point 

````python
# This is the url you will scrape in this exercise
url5 = 'http://uscode.house.gov/download/download.shtml'
````

````python
print(url5)
````

````python
resp5=requests.get(url5)
html5=resp5.text
soup5=BeautifulSoup(html5,'html.parser')
soup5
titles_changed = soup5.findAll('div', attrs={'class':'usctitlechanged'})
titles_changed
````

````python
titles_lst=[]
for i in titles_changed:
    titles_lst.append(i.text.replace('\n','').replace('٭','').strip())
titles_lst
````
