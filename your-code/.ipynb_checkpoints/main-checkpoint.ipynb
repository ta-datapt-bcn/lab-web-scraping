````python
#your code

lst = soup.find_all('h1', {'class': 'h3 lh-condensed'}) 

lst
````

````python


names = [name.get_text() for name in lst]

names
````

````python
#also try it with regex

lst_str = [str(i) for i in lst]
````

````python
lst_clean = [i[:-9] for i in lst_str]
lst_cleaned = [i[38:] for i in lst_clean]


lst_names = []
for i in lst_cleaned:
    m =re.search('\>(.*)', i)
    lst_names.append(m.group(0))

lst_names = [i[1:] for i in lst_names]
lst_names
````

````python
lst_id_clean = [i[38:] for i in lst_clean]
lst_id_names = []
for i in lst_id_clean:
    m = re.search('(.*)\>', i)
    lst_id_names.append(m.group(0))

lst_id_names = [i[:-2] for i in lst_id_names]
lst_id_names
````

````python
lst_full = []
for i in range(len(lst_names)):
    lst_full.append(lst_names[i] + " " + '(' + lst_id_names[i] + ')')
    
lst_full
````

#### Display all the image links from Walt Disney wikipedia page

````python
html3 = urlopen('https://en.wikipedia.org/wiki/Walt_Disney')
bs = BeautifulSoup(html3, 'lxml')
images = bs.find_all('img', {'src':re.compile('.jpg')})
for image in images: 
    print(image['src']+'\n')
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url4 ='https://en.wikipedia.org/wiki/Python' 
````

````python
#your code

html4 = urlopen('https://en.wikipedia.org/wiki/Monty_Python')
bs = BeautifulSoup(html4, 'lxml')
links = bs.find_all('a', attrs={'href': re.compile("^http://")})
for link in links: 
    print(link.get('href'))
````
